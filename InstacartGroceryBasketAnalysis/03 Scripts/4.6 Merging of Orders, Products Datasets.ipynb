```python
#Merging Orders Products Combined with Products
## 09 Using a left join to enrich "orders_products_combined" with matching data: product_name, aisle_id, dept_id and prices
## Using al alternative name to "_merge," "Check" so that an error message won't come up with indicator
df_merged = df_orders_products_combined.merge(df_prods, on = ['product_id'], how = 'left', indicator = 'Check')
```
