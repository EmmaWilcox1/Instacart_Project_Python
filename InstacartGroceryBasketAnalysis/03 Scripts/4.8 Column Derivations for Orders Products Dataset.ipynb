```python
#Changing datatypes
##11 Changing datatypes to smaller usage, using .astype
df_orders_products_merged['user_id'] = df_orders_products_merged['user_id'].astype('int64')
df_orders_products_merged['order_number']=df_orders_products_merged['order_number'].astype('int64')
df_orders_products_merged['orders_day_of_week']=df_orders_products_merged['orders_day_of_week'].astype('int8')
df_orders_products_merged['time_order_placed']=df_orders_products_merged['time_order_placed'].astype('int8')
df_orders_products_merged['days_since_prior_order']=df_orders_products_merged['days_since_prior_order'].astype('float16')
df_orders_products_merged['add_to_cart_order']=df_orders_products_merged['add_to_cart_order'].astype('int8')
df_orders_products_merged['reordered']=df_orders_products_merged['reordered'].astype('int8')
df_orders_products_merged['_merge']=df_orders_products_merged['_merge'].astype('category')
df_orders_products_merged['product_name']=df_orders_products_merged['product_name'].astype('object')
df_orders_products_merged['aisle_id']=df_orders_products_merged['aisle_id'].astype('float16')
df_orders_products_merged['department_id']=df_orders_products_merged['department_id'].astype('float16')
df_orders_products_merged['Check']=df_orders_products_merged['Check'].astype('category')
```

```python
##12 Using .info to confirm datatypes were changed
df_orders_products_merged.info()
```

```python
##13 Using .info to review data
df_orders_products_merged.head()
```

```python
#Creating new columns related to ordering and product costs
##14 Using .loc instead of a user defined function, to sort prices into three ranges, part 1
df_orders_products_merged.loc[df_orders_products_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
##15 Using .loc instead of a user defined function, to sort prices into three ranges, part 2
df_orders_products_merged.loc[(df_orders_products_merged['prices'] <= 15) & (df_orders_products_merged['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
##16 Using .loc instead of a user defined function, to sort prices into three ranges, part 3
df_orders_products_merged.loc[df_orders_products_merged['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
##17 ##15 Using .valuecounts to review range of prices
df_orders_products_merged['price_range_loc'].value_counts(dropna = False)
```

```python
##18 Client now wants to sort days into three ranges, 
##that include the 2 slowest days and the 2 busiest days, renaming 
## column "busiest_days"
result1 = []

for value in df_orders_products_merged["orders_day_of_week"]:
  if value in [0, 1]:
    result1.append("busiest_days")
  elif value in [3, 4]:
    result1.append("Slowest_days")
  else:
    result1.append("Regularly busy")
```

```python
##19 Creating a column, "busiest_days_column" and filling it with the 3 ranges
df_orders_products_merged['busiest_days_column'] = result1
```

```python
##20 Using .valuecounts to review range
df_orders_products_merged['busiest_days_column'].value_counts(dropna = False)
```

```python
##21 Using .valuecounts to review frequency of orders, by hour of the day
df_orders_products_merged['time_order_placed'].value_counts(dropna = False)
```

```python
##22 Client wants to create 3 ranges, allowing them to address an issue with website 
## freezing during busiest periods of each day's order. My method is to group the 24-hour day's frequency counts
## into 3 8-hour groups. 
result3 = []

for value in df_orders_products_merged["time_order_placed"]:
  if value in [10, 11, 14, 12, 16, 15, 13, 9]:
    result3.append("most_orders")
  elif value in [17, 18, 8, 19, 20, 7, 21, 22]:
    result3.append("average_orders")
  else:
    result3.append("fewest_orders")
```

```python
##23 Creating a column, "busiest_period_of_day" and filling it with the 3 ranges
df_orders_products_merged['busiest_period_of_day'] = result3
```

```python
##24 Using .valuecounts to look at range
df_orders_products_merged['busiest_period_of_day'].value_counts(dropna = False)
```

```python
##25 Using .groupby and .agg to get the average orders for each department
## The highest mean order number for the subset was Dept #18, bur for the overall dataset it is Dept #21.
df_orders_products_merged.groupby('department_id').agg({'order_number': ['mean']})
```

```python
##26 Get the max numbers of orders for each user, and put the results in a column, "max_order"
df_orders_products_merged['max_order'] = df_orders_products_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
##27 Using .value_counts to look at range of maximum number of orders range, there is alot of diversity
df_orders_products_merged['max_order'].value_counts(dropna = False)
```

```python
##28 Using .head to look at data in newer columns
df_orders_products_merged.head(100)
```

```python
##29 Part 1 of 3 steps using .loc to sort customers into 3 ranges, of loyal, regular and new
df_orders_products_merged.loc[df_orders_products_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
##30 Part 2 of 3 steps using .loc to sort customers into 3 ranges, of loyal, regular and new
df_orders_products_merged.loc[(df_orders_products_merged['max_order'] <= 40) & (df_orders_products_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
##31 Part 3 of 3 steps using .loc to sort customers into 3 ranges, of loyal, regular and new
df_orders_products_merged.loc[df_orders_products_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
##32 Using .head to review the newer data
df_orders_products_merged[['user_id', 'loyalty_flag', 'max_order', 'order_number']].tail(60)
```

```python
##33 using .value_counts to look at range of customer loyalty
df_orders_products_merged['loyalty_flag'].value_counts(dropna = False)
```

```python
##34 Using .groupby and .mean to get average price of items purchased by each user
df_orders_products_merged.groupby('user_id')['prices'].mean()
```

```python
##35 using .transform to create a new column, "mean_order" to show the average prices
df_orders_products_merged['mean_order'] = df_orders_products_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
##36 using .head to confirm contents of new column
df_orders_products_merged.head()
```

```python
##37  Using .loc to label those with average ordered product prices equal to or over 10 as High Spenders
df_orders_products_merged.loc[df_orders_products_merged['mean_order'] >= 10, 'spending_level'] = 'High Spender'
```

```python
##38 Using .loc to label those with average ordered product prices under 10 as Low Spenders
df_orders_products_merged.loc[df_orders_products_merged['mean_order'] < 10, 'spending_level'] = 'Low Spender'
```

```python
##39 Using .value_counts to look at range of spending styles
df_orders_products_merged['spending_level'].value_counts(dropna = False)
```

```python
##40 Using .groupby and .median to get the median number of days since a user last placed an order
df_orders_products_merged.groupby('user_id')['days_since_prior_order'].median()
```

```python
##41 Using .transform to create a new column, "median_order_time" for the new data
df_orders_products_merged['median_order_time'] = df_orders_products_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
##42 using .head to review contents of new column
df_orders_products_merged.head()
```

```python
##43 Using .loc to label those with median times of over 20 "Non-frequent customer"
df_orders_products_merged.loc[df_orders_products_merged['median_order_time'] > 20, 'frequency_of_customer'] = 'Non-Frequent Customer'
```

```python
##44 Using .loc to label those with median times of equal to or under 20, but over 10 days, "Regular Customer"
df_orders_products_merged.loc[(df_orders_products_merged['median_order_time'] <=20) & (df_orders_products_merged['median_order_time'] > 10), 'frequency_of_customer'] = 'Regular Customer'
```

```python
##45 Using .loc to label those with median times of equal to or under 10 "Frequent customer"
df_orders_products_merged.loc[df_orders_products_merged['median_order_time'] <=10, 'frequency_of_customer'] = 'Frequent Customer'
```

```python
##46 Using .value_counts to look at range
df_orders_products_merged['frequency_of_customer'].value_counts(dropna = False)
```
